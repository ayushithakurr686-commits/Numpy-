##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np
arr = np.array([10,20,30,40,50,60])

#Pick element at position 0,2,4
arr[[0,2,4]]
```

You can do this fancy indexing in 2D Matrix aswell

```python
matrix = np.arange(1,13).reshape(3,4)
#matrix[[row0,colm1],[row1,colm0]]
matrix[[0,1],[1,2]]
print(matrix)
matrix[[1,2],[0,2]]
```

##Conditional Selection
Find values based on conditions or replace them.

```python

arr =np.array([5,7,11,18,30])

#we will get the indices only
#which are greater than 10
np.where(arr > 10)

#array mai jaha bhi less than 10 values
#aa rhi hai waha 0 se replace ho jayegi
print(np.where(arr < 10,arr ,0))
```

```python
arr2 =np.array([23,25,30,50,100,75,80])

#replace the value withh 0 which are less than 55
np.where(arr2 < 55,arr2,0)
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
#aam zindagi
array = np.array([2,10,6,47,89])
print(np.sort(array))

index = np.argsort([2,10,6,47,89])
print(index)
```

```python
#Matrix sort
matrix = np.array([[4,3,12],[14,10,15]])
#1. Row wise sort(axis =0)
np.sort(matrix,axis =1)

#2. Column wise sort(axis=1)
np.sort(matrix,axis =0)

#3.Sort complete matrix
np.sort(matrix)
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.concatenate([a,b])
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[4,5]])
B = np.array([[7,8],[10,11]])

#Vertical Stacking
np.vstack([A,B])

#Horizontal Stacking
np.hstack([A,B])
```
